```python
%pylab inline
%matplotlib inline

import socket 
import struct
import array
import numpy
import abc
import matplotlib.pyplot as plt
import pylab as pl
import time
from random import randint
from IPython import display
from ITraderAgent import ITraderAgent
from TraderGameEnvironment import TraderGameEnvironment

class TraderPainter(ITraderAgent):
    
    def __init__(self): pass
    
    def plot_graph(self, dataMatrix):
        (rows, cols) = dataMatrix.shape
        x = numpy.arange(0, cols, 1)
        pl.clf()
        for i in [0, 1, 2, 3]: # <open close min max> for main instrument
            pl.plot(x, dataMatrix[i, :])
            
    def print_additional_info(self, dataMatrix):
        (rows, cols) = dataMatrix.shape
        print "---------------------------------"
        print "Balance:", dataMatrix[rows-1, cols-1], "$"
        print "Use of leverage:", dataMatrix[rows-3, cols-1]*100, "%"
        print "Opened orders:", dataMatrix[rows-4, cols-1]
        print "Time:", time.strftime("%a, %d %b %Y %H:%M:%S +0000", time.gmtime(dataMatrix[rows-2, cols-1]/1000.0))
        print "---------------------------------"

    def start_episode(self, observation):
        display.clear_output(wait=True)
        print "Start episode"
        action = randint(0, 2) - 1
        print "Action:", action
        self.plot_graph(observation)
        display.display(pl.gcf())
        
        return action

    def step(self, reward, observation):
        display.clear_output(wait=True)
        print "Reward=", reward
        action = randint(0, 2) - 1
        print "Action:", action
        self.print_additional_info(observation)
        self.plot_graph(observation)
        display.display(pl.gcf())
        
        return action

    def end_episode(self, reward, terminal=True):
        display.clear_output(wait=True)
        print "End episode"
        print "Won=", not terminal
        print "Reward"
        display.display(pl.gcf())

# MAIN point of entry <---
if __name__ == "__main__":
    agent = TraderPainter()
    game = TraderGameEnvironment(
        TraderGameEnvironment.HOST,
        TraderGameEnvironment.PORT,
        agent)
    game.play()
```
