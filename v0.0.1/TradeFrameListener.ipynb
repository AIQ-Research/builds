```python
%pylab inline
%matplotlib inline

import socket 
import struct
import array
import numpy
import abc
import matplotlib.pyplot as plt
import pylab as pl
import time

from IPython import display

host = '' 
port = 6666 
backlog = 5 
size = 8192 


def server_loop():

    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM) 
    s.bind((host,port)) 
    s.listen(backlog) 

    while True:

        client, address = s.accept() 

        display.clear_output(wait=True)

        print "from", address
        data = client.recv(size)
        print "len(data)", len(data)
        state = struct.unpack("d", data[0:8])[0]
        print "state:", state
        rows = struct.unpack("d", data[8:16])[0]
        cols = struct.unpack("d", data[16:24])[0]
        print rows, cols
        value = array.array('d', data[24:])
        dataList = value.tolist() 
        print "len(data):", len(value)
        if len(value) == rows*cols:
            dataMatrix = numpy.asarray(dataList)
            dataMatrix = numpy.reshape(dataMatrix, (rows, cols))
            #print dataMatrix 
            x = numpy.arange(0, cols, 1)

            pl.clf()
            #for i in range(0, int(rows)):
            for i in [0, 1]:
                pl.plot(x, dataMatrix[i, :])

            display.display(pl.gcf())
            time.sleep(0.1)

        else:
            print "Something goes wrong with the data transferring"

        array2send = [1.]

        client.send(array.array('d', array2send).tostring())
        client.close() 

    s.close()


server_loop()
```
